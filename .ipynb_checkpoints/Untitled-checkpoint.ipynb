```python
import numpy as np
import random
import os
```

```python
def initialize_string_matrix(lines):
  string_mat = []
  for line in lines:
    if line[0] == '0' or line[0] == '1':
      string_mat.append(line[:-3])
  return string_mat
```

```python
def initialize_ones_matrix(string_mat):
  matrix = []
  for string_row in string_mat:
    row = []
    for i in range(len(string_row)):
      if i % 2 == 0:
        row.append(int(string_row[i]))
    matrix.append(row)
  return matrix
```

```python
def read_matrix(file_name):
  with open(file_name, 'r') as f:
    lines = f.readlines()
  string_mat = initialize_string_matrix(lines)
  matrix = initialize_ones_matrix(string_mat)
  return np.array(matrix)
```

```python
M = np.array(["z" + str(i + 1) for i in range(matrix.shape[1])])
M
# M = np.array(['a', 'b', 'c', 'd', 'e', 'f'])
# M
```

```python
new_matrix = []
for row in matrix:
  new_matrix_row = []
  for row_index in range(len(row)):
    if row[row_index] == 1:
      new_matrix_row.append(M[row_index])
    else:
      new_matrix_row.append("")
  new_matrix.append(new_matrix_row)
new_matrix = np.array(new_matrix)
new_matrix
```

```python
indices = {}
index = 0
for letter in M:
  indices[letter] = index
  index += 1
indices
```

```python
N = []
for row in new_matrix:
  N_row = []
  for elem in row:
    if elem != '':
      N_row.append(elem)
  N.append(np.array(N_row))
N = np.array(N, dtype='object')
N
```

```python
def get_lambda_max(lambda_set, indices):
  if len(lambda_set) == 0:
    return -1
  last_letter = lambda_set[-1]
  return indices[last_letter]
```

```python
def check(gamma_set, N):
  subsets_number = len(N)
  c_sigma = ['0' for i in range(subsets_number)]
  sigma = {}
  for i in range(subsets_number):
    subset = N[i]
    for letter in gamma_set:
      sigma[letter] = 0
      for j in range(len(subset)):
        subset_letter = subset[j]
        if letter == subset_letter:
          if c_sigma[i] == '0':
            c_sigma[i] = letter
          else:
            c_sigma[i] = 'X'
  # print("Sigma: ", sigma)
  projection = {}
  for sigma_letter in c_sigma:
    if sigma_letter != 'X' and sigma_letter != '0':
      projection[sigma_letter] = 0
      projection[sigma_letter] += 1
  # print("projection: ", projection)
  if len(projection) < len(gamma_set):
    return "KO"
  elif "0" in c_sigma:
    return "OK"
  return "MHS"
```

```python
def output(gamma_set):
  out = "Found a MHS: "
  for letter in gamma_set:
    out += letter + ", "
  print(out[:-1])
```

```python
from tqdm.auto import tqdm
queue = [[]]
while len(queue) > 0:
  lambda_set = queue.pop(0)
  # print(lambda_set)
  lambda_max_index = get_lambda_max(lambda_set, indices)
  letters = [M[e_index] for e_index in range(lambda_max_index + 1, len(M))]
  print("Letters: ", letters)
  
  for letter in tqdm(letters):
    gamma_set = lambda_set[:]
    gamma_set.append(letter)
    # print("Gamma Set: ", gamma_set)

    result = check(gamma_set, N)

    if result == "OK" and letter != M[-1]:
      # print("OK Set: ", gamma_set)
      queue.append(gamma_set)
    elif result == "MHS":
      output(gamma_set)
```
